```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import time

# Spotify API credentials
client_id = 'f5168c93172c4a74a4203211925e5145'  # Replace with your client ID
client_secret = 'ad6302dd5b7844d68dc533261cd29cd2'  # Replace with your client secret

# Authenticate with Spotify
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))

# Function to fetch Spotify track link
def get_spotify_link(track_name, artist_name):
    try:
        # Search for the track
        results = sp.search(q=f'track:{track_name} artist:{artist_name}', type='track', limit=1)
        if results['tracks']['items']:
            track = results['tracks']['items'][0]
            return track['external_urls']['spotify']  # Get the Spotify link
        else:
            return None
    except Exception as e:
        print(f"Error fetching link for {track_name} by {artist_name}: {e}")
        return None

# Add song links to the dataset in batches to handle rate limits
def add_song_links_with_batches(dataset, batch_size=50, delay=60):
    """
    Adds Spotify links to the dataset in batches to avoid hitting rate limits.

    Parameters:
        dataset (DataFrame): The dataset containing 'track_name' and 'artist_name' columns.
        batch_size (int): The number of rows to process per batch.
        delay (int): The delay (in seconds) between batches.
    """
    # Ensure the spotify_link column exists
    if 'spotify_link' not in dataset.columns:
        dataset['spotify_link'] = None

    # Process the dataset in batches
    for start in range(0, len(dataset), batch_size):
        end = start + batch_size
        print(f"Processing rows {start} to {end}...")

        for i in range(start, min(end, len(dataset))):
            try:
                if pd.isnull(dataset.at[i, 'spotify_link']):  # Skip rows with existing links
                    dataset.at[i, 'spotify_link'] = get_spotify_link(
                        dataset.at[i, 'track_name'], dataset.at[i, 'artist_name']
                    )
            except Exception as e:
                print(f"Error processing row {i}: {e}")

        # Save progress after each batch
        dataset.to_csv("SPOTIFY_DATASET_3.csv", index=False)
        print(f"Batch {start} to {end} saved. Waiting for {delay} seconds...")
        time.sleep(delay)  # Wait to avoid hitting the rate limit

    return dataset

# Load the dataset
try:
    df = pd.read_csv("SPOTIFY_DATASET_2.csv")
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: 'SPOTIFY_DATASET_2.csv' not found. Please ensure the file exists in the working directory.")
    exit()

# Check for required columns
if 'track_name' not in df.columns or 'artist_name' not in df.columns:
    print("Error: The dataset must contain 'track_name' and 'artist_name' columns.")
    exit()

# Remove duplicate entries to avoid redundant API requests
df = df.drop_duplicates(subset=['track_name', 'artist_name'])
print("Duplicates removed from the dataset.")

# Initialize df_with_links as df
df_with_links = df.copy()

# Add Spotify links to the dataset
try:
    print("Adding Spotify links in batches. This may take some time...")
    df_with_links = add_song_links_with_batches(df_with_links, batch_size=50, delay=60)
    print("Spotify links added successfully.")
except Exception as e:
    print(f"An error occurred while adding Spotify links: {e}")

# Save the updated dataset
try:
    df_with_links.to_csv("SPOTIFY_DATASET_3.csv", index=False)
    print("Updated dataset saved as 'SPOTIFY_DATASET_3.csv'.")
except Exception as e:
    print(f"An error occurred while saving the updated dataset: {e}")
```
